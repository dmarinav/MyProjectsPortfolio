```python
import matplotlib.pyplot as plt

def bar_chart(left_edges,heights,bar_width,color):
    return plt.bar (left_edges,heights,bar_width,color=colors)

def bar_title (title2):
    return plt.title(title2)

def bar_x_label (xlabel2):
    return plt.xlabel(xlabel2)

def bar_y_label (ylabel2):
    return plt.ylabel(ylabel2)

def bar_xticks(x_coords, xticks2):
    return plt.xticks(left_edges, xticks2)

def bar_yticks(y_coords, yticks2):
    return plt.yticks(heights, yticks2)
```

```python
import matplotlib.pyplot as plt


def line_chart(x_coords,y_coords, color1):
    return plt.plot(x_coords, y_coords, color1)

def line_title (title1):
    return plt.title(title1)

def line_x_label (xlabel1):
    return plt.xlabel(xlabel1)

def line_y_label (ylabel1):
    return plt.ylabel(ylabel1)

def line_xticks(x_coords, xticks1):
    return plt.xticks(x_coords, xticks1)

def line_yticks(y_coords, yticks1):
    return plt.yticks(y_coords, yticks1)
```

```python
import matplotlib.pyplot as plt

def pie_chart(slices, labels, colors,autopct):
        return plt.pie (slices,labels=lab, colors=col,autopct='%1.1f%%')

def pie_title (title3):
    return plt.title(title3)
```

```python
import matplotlib.pyplot as plt

LINE_CHART_CHOICE = 1
BAR_CHART_CHOICE = 2
PIE_CHART_CHOICE = 3
QUIT_CHOICE = 4


def display_menu():
    print('MENU:')
    print('1 Line Chart')
    print('2 Bar Chart')
    print('3 Pie Chart')
    print('4 Quit')
    print()
    
choice = 0

while choice != QUIT_CHOICE:
    display_menu()
    choice = int(input('Enter your choice: '))
    print()

    if choice == BAR_CHART_CHOICE:
        question= str(input("Do you have a file with data to open? "))
        print()
        if question=='Yes' or question =='yes':
            try:
                open_file=str(input("Please enter the name of the file you would like to use: "))
                print()
                with open(open_file) as bar_file:
                    line1 = bar_file.readline().strip('\n').strip('[]')
                    left_edges=[int(x) for x in line1.split(',')]
                    line2 = bar_file.readline().strip('\n').strip('[]')
                    heights=[int(x) for x in line2.split(',')]
                    line3 = bar_file.readline().strip('\n').strip('[]')
                    bar_width=float(line3)
                    line4 = bar_file.readline().strip('\n').strip('[]')
                    line4x=line4.replace("'", "")
                    colors=[str(x) for x in line4x.split(', ')]
                    title2=bar_file.readline().strip('\n')
                    xlabel2=bar_file.readline().strip('\n')
                    ylabel2=bar_file.readline().strip('\n')
                    line8 = bar_file.readline().strip('\n').strip('[]')
                    line8x=line8.replace("'", "")
                    xticks2=[str(x) for x in line8x.split(', ')]
                    line9 = bar_file.readline().strip('\n').strip('[]')
                    line9x=line9.replace("'", "")
                    yticks2=[str(x) for x in line9x.split(', ')]
                    bar_chart(left_edges,heights,bar_width, color=colors)
                    bar_title(title2)
                    bar_x_label(xlabel2)
                    bar_y_label(ylabel2)
                    bar_xticks(left_edges, xticks2)
                    bar_yticks(heights, yticks2)
                    plt.show()
                    pass
            except IOError:
                print('The filename', open_file, 'does not exit. Please repeat your selection!')
                print()
                pass
            
        elif question=='No' or question =='no':         
            while True:
                try:
                    left_edges  = [int(x) for x in input("Enter left edges separated by comma: ").split(',')]
                    heights = [int(x) for x in input("Enter heights separated by comma: ").split(',')]
                    bar_width = float(input("Enter bar width: "))
                except ValueError:
                    print()
                    print('Please enter numbers only!')
                    print()
                else:
                    colors=[str(x.strip()) for x in input("Enter colors separated by comma:").split(',')]
                    title2 = str(input("Enter your title: "))
                    xlabel2=str(input("Enter x-axis label separated by comma: "))
                    ylabel2=str(input("Enter y-axis label separated by comma: "))
                    xticks2=[str(x.strip()) for x in input("Enter x ticks separated by comma:").split(',')]
                    yticks2=[str(x.strip()) for x in input("Enter y ticks separated by comma:").split(',')]
                    print()
                    save_file = str(input("Would you like to save the input data? "))
                    print()
                    if save_file =="yes" or save_file =="Yes":
                        name_file=str(input("Enter the name of the file: "))
                        file = open(name_file, 'w')
                        file.write(str(left_edges)+'\n')
                        file.write(str(heights)+'\n')
                        file.write(str(bar_width)+'\n')
                        file.write(str(colors) +'\n')
                        file.write(str(title2)+'\n')
                        file.write(str(xlabel2)+'\n')
                        file.write(str(ylabel2)+'\n')
                        file.write(str(xticks2)+'\n')
                        file.write(str(yticks2)+'\n')
                        file.close()
                        print()
                        print ("The input data has been saved to",file)
                        print()
                        
                    if len(left_edges)==len(heights)==len(colors)==len(xticks2)==len(yticks2):
                        bar_chart(left_edges,heights,bar_width, color=colors)
                        bar_title(title2)
                        bar_x_label(xlabel2)
                        bar_y_label(ylabel2)
                        bar_xticks(left_edges, xticks2)
                        bar_yticks(heights, yticks2)
                        plt.show()
                        break
                        pass
                    else:
                        print()
                        print('The number of entered data points does not match:')
                        print()
                        print('# of edges entered = ', len(left_edges))
                        print('# of heights entered = ',len(heights))
                        print('# of colors entered = ', len(colors))
                        print('# of x ticks entered = ', len(xticks2))
                        print('# of y ticks entered = ', len(yticks2))
                        print()
                        print('Please enter your data again!')
                        print()
                        pass
        else:
            print('You have to enter yes or no. Please repeat your selection!')
            print()
  


    elif choice == LINE_CHART_CHOICE:
        question= str(input("Do you have a file with data to open? "))
        print()
        if question=='Yes' or question =='yes':
            try:
                open_file=str(input("Please enter the name of the file you would like to use: "))
                print()
                with open(open_file) as line_file:
                    line1 = line_file.readline().strip('\n').strip('[]')
                    x_coords=[int(x) for x in line1.split(',')]
                    line2 = line_file.readline().strip('\n').strip('[]')
                    y_coords=[int(x) for x in line2.split(',')]
                    line3 = line_file.readline().strip('\n')
                    color1=str(line3)
                    title1=line_file.readline().strip('\n')
                    xlabel1=line_file.readline().strip('\n')
                    ylabel1=line_file.readline().strip('\n')
                    line7 = line_file.readline().strip('\n').strip('[]')
                    line7x=line7.replace("'", "")
                    xticks1=[str(x) for x in line7x.split(', ')]
                    line8 = line_file.readline().strip('\n').strip('[]')
                    line8x=line8.replace("'", "")
                    yticks1=[str(x) for x in line8x.split(', ')]
                    line_chart(x_coords,y_coords, color1)
                    line_title(title1)
                    line_x_label(xlabel1)
                    line_y_label(ylabel1)
                    line_xticks(x_coords, xticks1)
                    line_yticks(y_coords, yticks1)
                    plt.show()
                    pass  
            except IOError:
                print('The filename', open_file, 'does not exit. Please repeat your selection!')
                print()
                pass
            
        elif question=='No' or question =='no':
            while True:
                try:
                    x_coords = [int(x) for x in input("Enter x_coords separated by comma: ").split(',')]
                    y_coords = [int(x) for x in input("Enter y_coords separated by comma: ").split(',')]
                except ValueError:
                    print()
                    print('Please enter numbers only!')
                    print()
                else:
                    color1 = str(input("Enter color of the line: "))
                    title1 = str(input("Enter your title: "))
                    xlabel1=str(input("Enter x-axis label: "))
                    ylabel1=str(input("Enter y-axis label: "))
                    xticks1=[str(x.strip()) for x in input("Enter x ticks separated by comma:").split(',')]
                    yticks1=[str(x.strip()) for x in input("Enter y ticks separated by comma:").split(',')]
                    print()
                    save_file = str(input("Would you like to save the input data? "))
                    print()
                    if save_file =="yes" or save_file =="Yes":
                        name_file=str(input("Enter the name of the file: "))
                        file = open(name_file, 'w')
                        file.write(str(x_coords)+'\n')
                        file.write(str(y_coords)+'\n')
                        file.write(str(color1)+'\n')
                        file.write(str(title1)+'\n')
                        file.write(str(xlabel1)+'\n')
                        file.write(str(ylabel1)+'\n')
                        file.write(str(xticks1)+'\n')
                        file.write(str(yticks1)+'\n')
                        file.close()
                        print()
                        print ("The input data has been saved to",file)
                        print()
                    if len(x_coords)==len(y_coords)==len(xticks1)==len(yticks1):
                        line_chart(x_coords,y_coords, color1)
                        line_title(title1)
                        line_x_label(xlabel1)
                        line_y_label(ylabel1)
                        line_xticks(x_coords, xticks1)
                        line_yticks(y_coords, yticks1)
                        plt.show()
                        break
                        pass
                    else:
                        print()
                        print('The number of entered data points does not match:')
                        print()
                        print('# of x_coords entered = ', len(x_coords))
                        print('# of y_coords entered = ',len(y_coords))
                        print('# of x ticks entered = ', len(xticks1))
                        print('# of y ticks entered = ', len(yticks1))
                        print()
                        print('Please enter your data again!')
                        print()
                        pass
        else:
            print('You have to enter yes or no. Please repeat your selection!')
            print()
            
    elif choice == PIE_CHART_CHOICE:
        question= str(input("Do you have a file with data to open? "))
        print()
        if question=='Yes' or question =='yes':
            try:
                open_file=str(input("Please enter the name of the file you would like to use: "))
                print()
                with open(open_file) as pie_file:
                    line1 = pie_file.readline().strip('\n').strip('[]')
                    slices=[float(x) for x in line1.split(',')]
                    line2 = pie_file.readline().strip('\n').strip('[]')
                    line2x=line2.replace("'", "")
                    lab=[str(x) for x in line2x.split(', ')]
                    line3 = pie_file.readline().strip('\n').strip('[]')
                    line3x=line3.replace("'", "")
                    col=[str(x) for x in line3x.split(', ')]
                    title3=pie_file.readline().strip('\n')
                    pie_chart(slices,labels=lab, colors=col,autopct='%1.1f%%')
                    pie_title(title3) 
                    plt.show()
                    pass
            except IOError:
                print()
                print('The filename', open_file, 'does not exit. Please repeat your selection!')
                print()
                pass 
            
        elif question=='No' or question =='no': 
            while True:
                try:
                    slices = [float(x) for x in input("Enter values of slices separated by comma: ").split(',')]
                except ValueError:
                    print()
                    print('Please enter numbers only!')
                    print()
                else:
                    lab = [str(x) for x in input("Enter labels separated by comma:").split(',')]
                    col=[str(x.strip()) for x in input("Enter colors separated by comma:").split(',')]
                    title3 = str(input("Enter your title: "))
                    print()
                    save_file = str(input("Would you like to save the input data? "))
                    print()
                    if save_file =="yes" or save_file =="Yes":
                        name_file=str(input("Enter the name of the file: "))
                        file = open(name_file, 'w')
                        file.write(str(slices)+'\n')
                        file.write(str(lab)+'\n')
                        file.write(str(col)+'\n')
                        file.write(str(title3)+'\n') 
                        file.close()
                        print()
                        print ("The input data has been saved to",file) 
                        print()

                    if len(slices)==len(lab)==len(col):
                        pie_chart(slices,labels=lab, colors=col,autopct='%1.1f%%')
                        pie_title(title3) 
                        plt.show()
                        break
                        pass
                    else:
                        print()
                        print('The number of entered data points does not match:')
                        print()
                        print('# of slices entered = ', len(slices))
                        print('# of labels entered = ',len(lab))
                        print('# of colors entered = ', len(col))
                        print()
                        print('Please enter your data again!')
                        print()
                        pass
        else:
            print('You have to enter yes or no. Please repeat your selection!')
            print()
            
    elif choice == QUIT_CHOICE:
        print('Existing program...')
        print()

    else:
        print('Invalid input! Please enter 1 to 4.')
        print()


    
```
